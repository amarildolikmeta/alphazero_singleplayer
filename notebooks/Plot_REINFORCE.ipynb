```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pits = pd.read_csv("RaceStrategy/REINFORCE_500/pits.csv")
loss = pd.read_csv("RaceStrategy/REINFORCE_500/loss.csv")
ret = pd.read_csv("RaceStrategy/REINFORCE_500/return.csv")
```

```python
sns.lineplot(x="iteration", y="n_pits", data=pits)
```

```python
sns.lineplot(x="iteration", y="loss", data=loss)
```

```python
sns.lineplot(x="iteration", y="return", data=ret)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)

fig.set_size_inches(18.5, 10.5)

sns.lineplot(x="iteration", y="n_pits", data=pits, ax=ax1)
sns.lineplot(x="iteration", y="loss", data=loss, ax=ax2)
sns.lineplot(x="iteration", y="return", data=ret, ax=ax3)

ax1.set_xlabel("")  # remove y label, but keep ticks
ax2.set_xlabel("")  # remove y label, but keep ticks

fig.tight_layout(pad=3.0)
fig.suptitle("REINFORCE performance")

plt.show()
```
