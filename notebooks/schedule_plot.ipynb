```python
import matplotlib.pyplot as plt
from tqdm import trange
```

```python
def schedule(x, k=1, width=1, min_depth=1) -> float:
    """The method computes a reduction factor for the budget to use during search.
    The intuition is that it requires less budget to build the full tree when next to the final states,
    so the search would take lots of time.

    """

    if x <= min_depth:
        return 1
    elif x == width:
        return 0

    x -= min_depth

    width = float(width)
    norm_x = x / width
    parenth = norm_x / (1 - norm_x)
    denom = 1 + parenth ** k
    return 1 / denom
```

```python
values = []
for i in trange(70):
    values.append(max(schedule(i, k=4.8, min_depth=15, width=70) * 50000, 10000))
plt.figure()
plt.plot(values)
plt.savefig("../figures/scheduler_profile")
plt.show()
```
